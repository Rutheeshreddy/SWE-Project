```python
from faker import Faker
import random
import string

# Initialize Faker to generate fake data
fake = Faker()

# Generate login data
logins = []
roles = ['student', 'instructor', 'admin']
for _ in range(100):  # Generate 100 login entries
    username = fake.user_name()
    password = ''.join(random.choices(string.ascii_letters + string.digits, k=8))  # Generate a random password
    role = random.choice(roles)
    logins.append((username, password, role))

# Generate student data
branch_code = ['cs','ai','ee','ma']
students = []
for _ in range(50):  # Generate 50 student entries
    branch = random.choice(branch_code)
    year = random.choice(['21','22'])
    id = 'btech' + ''.join(random.choices(string.ascii_letters + string.digits, k=8))
    name = fake.name()
    department = branch_code
    year = '20' + year
    students.append((id, name, department))

# Generate instructor data
instructors = []
for _ in range(20):  # Generate 20 instructor entries
    id = fake.uuid4()
    name = fake.name()
    department = fake.word(ext_word_list=['Math', 'Physics', 'Chemistry', 'Biology'])
    instructors.append((id, name, department))

# Generate past courses data
past_courses = []
for _ in range(50):  # Generate 50 past course entries
    course_id = fake.random_int(min=1000, max=9999)
    semester = random.choice(['Spring', 'Summer', 'Fall'])
    year = fake.random_int(min=2010, max=2022)
    name = fake.word(ext_word_list=['Mathematics', 'Physics', 'Chemistry', 'Biology']) + ' ' + fake.word(ext_word_list=['101', '201', '301'])
    credits = fake.random_int(min=1, max=4)
    instructor_id = random.choice(instructors)[0]
    prerequisites = fake.sentence()
    max_capacity = fake.random_int(min=20, max=50)
    past_courses.append((course_id, semester, year, name, credits, instructor_id, prerequisites, max_capacity))

# Generate present courses data
present_courses = []
for _ in range(50):  # Generate 50 present course entries
    course_id = fake.random_int(min=1000, max=9999)
    semester = random.choice(['Spring', 'Summer', 'Fall'])
    year = fake.random_int(min=2023, max=2024)
    name = fake.word(ext_word_list=['Mathematics', 'Physics', 'Chemistry', 'Biology']) + ' ' + fake.word(ext_word_list=['101', '201', '301'])
    credits = fake.random_int(min=1, max=4)
    instructor_id = random.choice(instructors)[0]
    prerequisites = fake.sentence()
    slot = random.choice(['A', 'B', 'C', 'D'])
    max_capacity = fake.random_int(min=20, max=50)
    present_courses.append((course_id, semester, year, name, credits, instructor_id, prerequisites, slot, max_capacity))

# Generate student courses data
student_courses = []
for _ in range(200):  # Generate 200 student course entries
    course_id = fake.random_int(min=1000, max=9999)
    semester = random.choice(['Spring', 'Summer', 'Fall'])
    year = fake.random_int(min=2015, max=2024)
    name = fake.word(ext_word_list=['Mathematics', 'Physics', 'Chemistry', 'Biology']) + ' ' + fake.word(ext_word_list=['101', '201', '301'])
    credits = fake.random_int(min=1, max=4)
    student_id = random.choice(students)[0]
    grade = random.choice(['A', 'B', 'C', 'D', 'F',''])
    feedback = fake.sentence()
    student_courses.append((course_id, semester, year, name, credits, student_id, grade, feedback))

# Print the generated data
print("Logins:", logins[:5])
print("Students:", students[:5])
print("Instructors:", instructors[:5])
print("Past Courses:", past_courses[:5])
print("Present Courses:", present_courses[:5])
print("Student Courses:", student_courses[:5])
```
